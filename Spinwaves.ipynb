```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants for the dispersion relation
gamma = 2.21e5  # Gyromagnetic ratio in m/(A*s)
mu_0 = 4 * np.pi * 1e-7  # Permeability of free space in H/m
H_eff = 0.1  # Effective magnetic field in Tesla
D = 1e-4  # Exchange stiffness constant in m^2/s

# Create a range for the wave vector k
k = np.linspace(0, 1e7, 1000)  # Wave vector in 1/m

# Calculate the dispersion relation for the spin waves
omega = gamma * mu_0 * (H_eff + D * k**2)  # Frequency in rad/s

# Convert frequency from rad/s to GHz for better visualization
omega_GHz = omega / 1e9

# Plot the dispersion relation
plt.figure(figsize=(8,6))
plt.plot(k, omega_GHz, label=r'$\omega(k) = \gamma \mu_0 (H_{eff} + Dk^2)$', color='b')
plt.xlabel(r'Wave vector $k$ (1/m)', fontsize=12)
plt.ylabel(r'Frequency $\omega$ (GHz)', fontsize=12)
plt.title('Dispersion Relation of Spin Waves', fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
